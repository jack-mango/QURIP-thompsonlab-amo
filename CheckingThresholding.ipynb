```python
import numpy as np

def get_pixel_centers(image_data):
    height, width = image_data.shape
    
    # Create coordinate grids using meshgrid
    x_coords, y_coords = np.meshgrid(np.arange(width), np.arange(height))
    
    # Calculate the center coordinates
    x_centers = x_coords + 0.5
    y_centers = y_coords + 0.5
    
    return x_centers, y_centers
```

```python
image_data = np.zeros((3, 3))

x_centers, y_centers = get_pixel_centers(image_data)

print("X Centers:")
print(x_centers)
print("Y Centers:")
print(y_centers)
```
