```python
fig,axs = plt.subplots(3,1,sharex=True)

idx = [0,2]

pixel_thresh = 1050
hot_thresh = 2

img = np.transpose(np.sum(data['nuvuImage'][idx[0]][idx[1]],axis=())[100:,:15])
occ = data['occupancy'][idx[0],:,idx[1]]
img_thresh = img > pixel_thresh

centers = np.array([(ii*5.9+6.9,7.5-ii*0.08) for ii in range(15)])

idx_filled = np.argwhere(occ).flatten()

scores = np.zeros(len(occ))

for pt in np.argwhere(img_thresh):
    #pt = np.argwhere(img_thresh)[0]

    dists = [np.linalg.norm(x) for x in centers - [pt[1],pt[0]]]
    idx_closest = np.argmin(dists)

    scores[idx_closest] += 1 if dists[idx_closest] < 2.8 else 0


occ_guess = scores >= hot_thresh
correct = np.argwhere([x == y for x,y in zip(occ_guess,occ)]).flatten()
false_positives = np.argwhere([x and (y==0) for x,y in zip(occ_guess,occ)]).flatten()
false_negatives = np.argwhere([(not x) and (y==1) for x,y in zip(occ_guess,occ)]).flatten()

# correct = np.argwhere((scores >= hot_thresh) == (occ == 1)).flatten()
# false_positives = np.argwhere((scores >= hot_thresh) == (occ == 0)).flatten()
# false_negatives = np.argwhere((scores < hot_thresh) == (occ == 1)).flatten()

axs[0].imshow(avg)
axs[1].imshow(img)
axs[2].imshow(img_thresh)

axs[0].plot([x[0] for x in centers],[x[1] for x in centers],'ws',alpha=1,fillstyle='none')
axs[1].plot([x[0] for x in centers],[x[1] for x in centers],'ks',alpha=1,fillstyle='none')
axs[1].plot([x[0] for x in centers[idx_filled]],[x[1] for x in centers[idx_filled]],'ws',alpha=1,fillstyle='none')

axs[2].plot([x[0] for x in centers[correct]],[x[1] for x in centers[correct]],'go',alpha=1,fillstyle='none',ms=20)
axs[2].plot([x[0] for x in centers[false_positives]],[x[1] for x in centers[false_positives]],'wo',alpha=1,fillstyle='none',ms=20)
axs[2].plot([x[0] for x in centers[false_negatives]],[x[1] for x in centers[false_negatives]],'ro',alpha=1,fillstyle='none',ms=20)



plt.tight_layout()
```

```python
dims = np.shape(data['nuvuImage'])[0:2]

res = []

pixel_thresh = 1050
hot_thresh = 2
centers = np.array([(ii*5.9+6.9,7.5-ii*0.08) for ii in range(15)])
    
for r in tqdm.tqdm(range(dims[0])):
#for r in tqdm.tqdm([3]):
    
    rundat = data['nuvuImage'][r]
    runocc = data['occupancy'][r]
    
    for s in range(dims[1]):
        
        idx = [r,s]

        #img = np.transpose(data['nuvuImage'][idx[0]][idx[1]][100:,:15])
        img = np.transpose(rundat[s][100:,:15])
        occ = runocc[:,s]
        img_thresh = img > pixel_thresh

        idx_filled = np.argwhere(occ).flatten()

        scores = np.zeros(len(occ))

        for pt in np.argwhere(img_thresh):
            #pt = np.argwhere(img_thresh)[0]

            dists = [np.linalg.norm(x) for x in centers - [pt[1],pt[0]]]
            idx_closest = np.argmin(dists)

            #scores[idx_closest] += 1 if dists[idx_closest] < 2.8 else 0
            scores[idx_closest] += np.exp(-(dists[idx_closest]/3)**2)

        occ_guess = scores >= hot_thresh
        correct = np.argwhere([x == y for x,y in zip(occ_guess,occ)]).flatten()
        false_positives = np.argwhere([x and (y==0) for x,y in zip(occ_guess,occ)]).flatten()
        false_negatives = np.argwhere([(not x) and (y==1) for x,y in zip(occ_guess,occ)]).flatten()

        res.append({'idx': [r,s], 'pixel_thresh': pixel_thresh, 'hot_thresh': hot_thresh,
                    'num': len(occ), 'occ': occ, 'occ_guess': occ_guess,
                    'correct': correct, 'false_positives': false_positives, 'false_negatives': false_negatives})
```

```python
ntot = nc = np.sum([x['num'] for x in res])
nc = np.sum([len(x['correct']) for x in res])
nfp = np.sum([len(x['false_positives']) for x in res])
nfn = np.sum([len(x['false_negatives']) for x in res])

nc/ntot,nfp/ntot,nfn/ntot
```
